## Assignment 2: Working with Pandas

Objective: Learn DataFrame creation and manipulation.

```python
import pandas as pd
```

### Creating DataFrame

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
df
```

### Summary Statistics

```python
df[['Age', 'Salary']].describe()
```

### Average HR Salary

```python
df[df['Department']=='HR']['Salary'].mean()
```

### Bonus Column

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

### Filtering Age Between 25 and 30

```python
df[(df['Age'] >= 25) & (df['Age'] <= 30)]
```

### Group By Department

```python
df.groupby('Department')['Salary'].mean()
```

### Sort and Save

```python
df.sort_values(by='Salary').to_csv('sorted_salary.csv', index=False)
```
