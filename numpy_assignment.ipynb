## Assignment 1: Working with NumPy

Objective: Understand NumPy basics, array creation, manipulation, and operations.

```python
import numpy as np
```

### 1. 1D Array Operations

```python
array_1d = np.arange(1, 21)

np.sum(array_1d), np.mean(array_1d), np.median(array_1d), np.std(array_1d), np.where(array_1d > 10)
```

### 2. 2D Array Operations

```python
array_2d = np.arange(1, 17).reshape(4, 4)
array_2d, array_2d.T, np.sum(array_2d, axis=1), np.sum(array_2d, axis=0)
```

### 3. Random Arrays

```python
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))

a, b, a+b, a-b, a*b, np.dot(a, b)
```

### 4. Reshaping and Slicing

```python
arr = np.arange(1, 13).reshape(3, 4)
arr, arr[:2, -2:]
```
